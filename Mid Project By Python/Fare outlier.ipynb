```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Dataset_midterm_Section(B).csv')
df.head(10)
```

```python
plt.figure(figsize=(5.5,3))
df.age.hist(bins=20,rwidth=0.7)
plt.ylabel('Count')
plt.xlabel('fare')
plt.show
```

```python
df.describe()
```

```python
df['fare'].plot.box(color = dict(boxes='DarkGreen', whiskers='DarkOrange',medians='DarkBlue', caps='Gray'),sym='r+',figsize=(4,3)) 
```

```python
Q1 = df.fare.quantile(0.25)
Q3 = df.fare.quantile(0.75)
Q1, Q3
```

```python
IQR = Q3 - Q1
IQR
```

```python
lower_limit = Q1 - 1.5*IQR
upper_limit = Q3 + 1.5*IQR
lower_limit, upper_limit
```

```python
out = df[(df.fare<lower_limit)|(df.fare>upper_limit)]
out
```

```python
df2 = df[(df.fare>=lower_limit)&(df.fare<=upper_limit)]
df2
```

```python
df2.shape
```

```python
df2['fare'].plot.box(color = dict(boxes='DarkGreen', whiskers='DarkOrange',medians='DarkBlue', caps='Gray'),sym='r+',figsize=(5,3)) 
```
